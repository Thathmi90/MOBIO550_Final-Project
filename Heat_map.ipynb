```python
# STEP 3: Load data
file_name = list(uploaded.keys())[0]
df = pd.read_excel(file_name, sheet_name='Sheet1')
```

```python
df.dropna(inplace = True)
print(df.shape)
```

```python
df = df.set_index('Protein IDs')
```

```python
sns.clustermap(df, metric="euclidean", z_score=0, method="average",cmap="viridis")
png_path = 'HeatMAp.png'
pdf_path = 'HeatMAp.pdf'
plt.savefig(png_path, dpi=600)
plt.savefig(pdf_path)
plt.show()
```

```python
sns.clustermap(df, metric="Correlation", z_score=0, method="average", cmap="icefire")
plt.show()
```
